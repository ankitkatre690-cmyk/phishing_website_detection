```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Numerical Features:", len(numerical_cols))
print("Categorical Features:", len(categorical_cols))
```

```python
numerical_summary = df[numerical_cols].describe().T
numerical_summary
```

```python
numerical_summary[['mean', '50%', 'std', 'min', 'max']]
```

```python
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
    print("Mode:", df[col].mode()[0])
```

```python
df[numerical_cols].hist(figsize=(20, 20), bins=30)
plt.suptitle("Histogram of Numerical Features", fontsize=16)
plt.show()
```

```python
# Select limited features to avoid clutter
selected_features = numerical_cols[:6]

sns.pairplot(df[selected_features])
plt.show()
```

```python

sns.pairplot(df[['length_url','domain_age','web_traffic','status']], hue='status')
```

```python
plt.figure(figsize=(18, 14))
corr_matrix = df[numerical_cols].corr()

sns.heatmap(corr_matrix, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
df['status'].value_counts()
```

```python
sns.countplot(x='status', data=df)
plt.title("Target Variable Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='length_url', hue='status', kde=True, bins=50)
plt.title('Distribution of URL Length by Status')
plt.xlim(0, 300) # Limiting to show main distribution
plt.savefig('url_length_dist.png')
```
